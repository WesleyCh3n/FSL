```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
epoch = 50
step = 525
tick = 5
```

```python
fig_title = "MobileNetV2(softmax) Ds Test Accuracy"
csvFile = "./exp/0315cross_Ds14_2400/logs/run-logs_20210315-183301_train-tag-test_accuracy.csv"
df = pd.read_csv(csvFile)
ax = df.plot(x="Step", y="Value", title=fig_title,
             legend=False, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0)
ax.set_xlabel("Epoch")
ax.set_xticklabels([i*tick for i in range(int(epoch/tick))])
ax
```

```python
fig_title = "MobileNetV2(softmax) Ds Test Loss"
csvFile = "./exp/0315cross_Ds14_2400/logs/run-logs_20210315-183301_train-tag-test_loss.csv"
df = pd.read_csv(csvFile)
ax = df.plot(x="Step", y="Value", title=fig_title,
             legend=False, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0)
ax.set_xlabel("Epoch")
ax.set_xticklabels([i*tick for i in range(int(epoch/tick))])
ax
```

```python
fig_title = "MobileNetV2(softmax) Ds Train Accuracy"
csvFile = "./exp/0315cross_Ds14_2400/logs/run-logs_20210315-183301_train-tag-train_accuracy.csv"
df = pd.read_csv(csvFile)
ax = df.plot(x="Step", y="Value", title=fig_title,
             legend=False, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0)
ax.set_xlabel("Epoch")
ax.set_xticklabels([i*tick for i in range(int(epoch/tick))])
ax
```

```python
fig_title = "MobileNetV2(softmax) Ds Train Loss"
csvFile = "./exp/0315cross_Ds14_2400/logs/run-logs_20210315-183301_train-tag-train_loss.csv"
df = pd.read_csv(csvFile)
ax = df.plot(x="Step", y="Value", title=fig_title,
             legend=False, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0)
ax.set_xlabel("Epoch")
ax.set_xticklabels([i*tick for i in range(int(epoch/tick))])
ax
```
