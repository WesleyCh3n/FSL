```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
epoch = 100
step = 240
tick = 5
```

```python
fig_title = "MobileNetV2 triplet loss in source domain (Ds)"
csvFile = "./run-logs_20210315-204341_train-tag-loss.csv"
df = pd.read_csv(csvFile)
ax = df.plot(x="Step", y="Value", title=fig_title, color='royalblue', 
             legend=False, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0)
plt.grid(True,axis='y')
ax.set_xlabel("Epoch")
ax.set_xticklabels([i*tick for i in range(int(epoch/tick))])
ax
```

```python
fig_title = "MobileNetV2 hardest distance in source domain (Ds)"
csvFile = "./run-logs_20210315-204341_train-tag-hardest_positive_dist.csv"
df = pd.read_csv(csvFile)
fig, ax = plt.subplots()
df.plot(x="Step", y="Value", label='hardest positive distance', title=fig_title, color='darkorange',
        legend=True, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0, ax=ax)

csvFile = "./run-logs_20210315-204341_train-tag-hardest_negative_dist.csv"
df = pd.read_csv(csvFile)
df.plot(x="Step", y="Value", label='hardest negative distance', title=fig_title, color='royalblue',
        legend=True, xticks=[i*step*tick for i in range(int(epoch/tick))], rot=0, ax=ax)
plt.grid(True,axis='y')
ax.set_xlabel("Epoch")
ax.set_xticklabels([i*tick for i in range(int(epoch/tick))])
ax
```
